### Get tokens, types and TTR

```python
import nltk
import urllib
from collections import Counter
#nltk.download('punkt')

webFile = urllib.request.urlopen("http://www.gutenberg.org/files/74/74-0.txt").read().decode("UTF-8")
webFile_lines = urllib.request.urlopen("http://www.gutenberg.org/files/74/74-0.txt").readlines()

tokens = nltk.word_tokenize(webFile)
types = Counter(tokens)
#print(tokens)

ttr = len(types)/len(tokens)
print("Number of Tokens: %d\nNumber of Types: %d\nTTR: %f" % (len(tokens),len(types), ttr))
```

### Type List

```python
print("Type\tFrequency")
print("----\t---------")
for token in sorted(types, key=types.get, reverse=True):
    print(token, "\t", types[token])
```

### Zipf's Law: Frequency vs Rank

```python
rank_list = sorted(list(set(types.values())), reverse=True)
i = 0
print("Rank\tFrequency")
print("----\t---------")
for val in rank_list:
    i += 1
    print(i,"\t",rank_list[i-1])
```

```python
import matplotlib.pylab as plt

x = list(range(1, i+1))
y = rank_list
plt.plot(x, y)
plt.show()
```

### Zipf's Law: Frequency vs Number of Meanings

```python
type_list = {"time":16, "long":13, "Project":8, "heart":5, "turned":6, "things":4, "people":5, "joy":3, "imperishable":1, "clay":2}
type_freq = {}

print("Number of Meanings\tFrequency\tToken")
print("------------------\t---------\t-----")
for token in sorted(type_list, key=type_list.get):
    print(type_list[token],"\t\t\t", types[token], "\t\t", token)
    type_freq[type_list[token]] = types[token]
```

```python
import matplotlib.pylab as plt

lists = sorted(type_freq.items())
x, y = zip(*lists)

plt.plot(x, y, "o")
plt.show()
```

### Zipf's Law: Frequency vs length

```python
length_freq = {}
i = 0
for token in types:
    #print(token, len(token), length_freq)
    #print(i)
    if len(token) in length_freq:
        i += 1
        length_freq[len(token)].append(types[token])
    else:
        
        length_freq[len(token)] = [types[token]]

print("Length\tFrequency")
print("------\t---------")
for length in sorted(length_freq):
    avg = sum(length_freq[length])/len(length_freq[length])
    length_freq[length] = avg
    print(length,"\t",length_freq[length])
```

```python
import matplotlib.pylab as plt

lists = sorted(length_freq.items())
x, y = zip(*lists)

plt.plot(x, y)
plt.show()
```

### Heap's Law

```python
token_size = []
vocab_size = []

# Growning at a rate of 200 line at a time
i = 0
line_set = ""
for line in webFile_lines:
    line = line.decode("UTF-8")
    if (i != 200):
        line_set += line
        i += 1
    else:
        i = 0
        token_size.append(len(nltk.word_tokenize(line_set)))
        vocab_size.append(len(set(nltk.word_tokenize(line_set))))

print("Tokens set Size\tVocabulary Size")
print("---------------\t---------------")
for val in zip(token_size,vocab_size):
    print(val[0], "\t\t", val[1])
```

```python
from scipy.optimize import curve_fit

def eqn(N,K,B):
   return K*(N**B)

p_opt,p_cov = curve_fit(eqn,token_size,vocab_size)
print("K \t\t\tBeta")
print("--\t\t\t-----")
print(str(p_opt[0]), "\t", str(p_opt[1]))

plt.figure(figsize=[15,10])
plt.plot(token_size,vocab_size, "o")
plt.plot(token_size,eqn(token_size,p_opt[0], p_opt[1] ))
plt.ylabel("Vocabulary Size")
plt.xlabel("Number of Tokens")
plt.title("Vocabulary vs Tokens")
plt.legend(["Datapoints", "Fitted Curve"])
plt.show()
```
